```python
from transformers import pipeline
```

```python
text_generator = pipeline("text-generation",model='EleutherAI/gpt-neo-1.3B')#We can run it with 2.7B too but I am using a smaller model over here
```

```python
topic = input('Write Down Your Esaay Topic') 
```

```python
res=text_generator(topic,max_length=200,do_sample=True,temperature=0.9)
```
